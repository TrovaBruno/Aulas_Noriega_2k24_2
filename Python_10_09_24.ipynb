```python
lista_notas=[3.4,4.5,7.0,8.0,5.5,7.5]
```

```python
soma=0
for nota in lista_notas:
    soma=soma+nota
    media =soma/len(lista_notas)
print(media)
```

```python
print('Media= ','{:.4f}'.format(media))
```
